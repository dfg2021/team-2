```python
import pandas as pd
import numpy as np

#Goal4 = pd.read_excel('Goal4.xlsx')
Goal4_csv = pd.read_csv('Goal4.csv', low_memory=False)

print(Goal4_csv.head())
#print(Goal4_csv)
```

```python
#dropping all the columns with only null values
Goal4_csv_new = Goal4_csv.dropna(how='all', axis=1)

#converting dataframe to the csv file
Goal4_csv_new.to_csv('Goal4_new.csv')
print(Goal4_csv_new)
```

```python
df = pd.DataFrame(Goal4_csv_new)
#print(df)

df['Value'] = df['Value'].fillna(df['Value'].mean())
#df['X2'] = df['X2'].fillna(df['X2'].median())

df['BasePeriod'] = df['BasePeriod'].fillna(df['BasePeriod'].median())

```
